```
import re
import string
from nltk.corpus import stopwords

def preprocess_data(df):
    # Remove special characters and punctuation
    df['headline'] = df['headline'].apply(lambda x: re.sub('[^a-zA-Z0-9\s]', '', x))
    
    # Convert all text to lowercase
    df['headline'] = df['headline'].apply(lambda x: x.lower())
    
    # Replace one or more digits with a whitespace - because digits don't matter here
    df['headline'] = df['headline'].str.replace('\d+', '')
    
    # Remove whitespace
    df['headline'] = df['headline'].str.strip()
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    df['headline'] = df['headline'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
    
    # Remove punctuation
    df['headline'] = df['headline'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
    
    return df
train_df = preprocess_data(train_df)
```

```
# make word cloud
import matplotlib.pyplot as plt
from wordcloud import WordCloud
# Define the categories in your dataset
categories = ['national', 'international', 'business & economy', 'sports', 'health']

# Loop through each category and generate a word cloud
for category in categories:
    # Extract the headlines for the current category
    headlines = train_df[train_df['category'] == category]['headline']
    
    # Concatenate all the headlines into a single string
    text = ' '.join(headlines)
    
    # Generate the word cloud
    wordcloud = WordCloud(width=800, height=800, background_color='white').generate(text)
    
    # Plot the word cloud
    plt.figure(figsize=(8,8))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.title(category)
    plt.show()
```

```
# Import necessary libraries
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Define a function to train and evaluate a model
def train_and_evaluate_model(X_train, y_train, X_test, y_test, model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('Accuracy:', accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))

# Split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(train_df['headline'], train_df['category'], test_size=0.2, random_state=42)

# Define a TF-IDF vectorizer
tfidf = TfidfVectorizer()

# Transform the train and test sets
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

# Define models
nb = MultinomialNB()
lr = LogisticRegression()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
svm = SVC()

# Define a list of models to train
models = [('Naive Bayes', nb),
          ('Logistic Regression', lr),
          ('Decision Tree', dt),
          ('Random Forest', rf),
          ('Support Vector Machines', svm)]

# Train and evaluate models
for model_name, model in models:
    print(model_name)
    train_and_evaluate_model(X_train_tfidf, y_train, X_test_tfidf, y_test, model)
```

```
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(y_test, y_pred):
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt='g', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Naive Bayes
nb_y_pred = nb.predict(X_test_tfidf)
print("Naive Bayes Confusion Matrix:")
plot_confusion_matrix(y_test, nb_y_pred)

# Logistic Regression
lr_y_pred = lr.predict(X_test_tfidf)
print("Logistic Regression Confusion Matrix:")
plot_confusion_matrix(y_test, lr_y_pred)

# Decision Tree
dt_y_pred = dt.predict(X_test_tfidf)
print("Decision Tree Confusion Matrix:")
plot_confusion_matrix(y_test, dt_y_pred)

# Random Forest
rf_y_pred = rf.predict(X_test_tfidf)
print("Random Forest Confusion Matrix:")
plot_confusion_matrix(y_test, rf_y_pred)

# Support Vector Machines
svm_y_pred = svm.predict(X_test_tfidf)
print("SVM Confusion Matrix:")
plot_confusion_matrix(y_test, svm_y_pred)
```

```
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# Evaluation metrics for different models
nb_scores = [accuracy_score(y_test, nb_y_pred),
             precision_score(y_test, nb_y_pred, average='weighted'),
             recall_score(y_test, nb_y_pred, average='weighted'),
             f1_score(y_test, nb_y_pred, average='weighted')]

lr_scores = [accuracy_score(y_test, lr_y_pred),
             precision_score(y_test, lr_y_pred, average='weighted'),
             recall_score(y_test, lr_y_pred, average='weighted'),
             f1_score(y_test, lr_y_pred, average='weighted')]

dt_scores = [accuracy_score(y_test, dt_y_pred),
             precision_score(y_test, dt_y_pred, average='weighted'),
             recall_score(y_test, dt_y_pred, average='weighted'),
             f1_score(y_test, dt_y_pred, average='weighted')]

rf_scores = [accuracy_score(y_test, rf_y_pred),
             precision_score(y_test, rf_y_pred, average='weighted'),
             recall_score(y_test, rf_y_pred, average='weighted'),
             f1_score(y_test, rf_y_pred, average='weighted')]

svm_scores = [accuracy_score(y_test, svm_y_pred),
              precision_score(y_test, svm_y_pred, average='weighted'),
              recall_score(y_test, svm_y_pred, average='weighted'),
              f1_score(y_test, svm_y_pred, average='weighted')]

# Create a dataframe to store the scores for each model
model_scores = pd.DataFrame({'Model': ['Naive Bayes', 'Logistic Regression', 'Decision Tree', 'Random Forest', 'SVM'],
                             'Accuracy': [nb_scores[0], lr_scores[0], dt_scores[0], rf_scores[0], svm_scores[0]],
                             'Precision': [nb_scores[1], lr_scores[1], dt_scores[1], rf_scores[1], svm_scores[1]],
                             'Recall': [nb_scores[2], lr_scores[2], dt_scores[2], rf_scores[2], svm_scores[2]],
                             'F1-Score': [nb_scores[3], lr_scores[3], dt_scores[3], rf_scores[3], svm_scores[3]]})

# Print the dataframe
print(model_scores)
```
